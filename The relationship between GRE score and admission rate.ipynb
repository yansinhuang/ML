```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import locale

# read csv
data = pd.read_csv('/Users/yan/Downloads/graduate-admissions/Admission_Predict.csv')

# Chinese 
mpl.rc('font', family='Arial Unicode MS')
```

```python
print(data.head())
```

```python
x = data['GRE Score']
y = data['Chance of Admit ']
plt.scatter(x,y)
```

非線性迴歸分析

```python
from sklearn.linear_model import LinearRegression
```

```python
X_poly = np.array([[k, k**2, k**3, k**4, k**5,  k**6] for k in x])
regr_poly = LinearRegression()
regr_poly.fit(X_poly, y)
Y_poly = regr_poly.predict(X_poly)
```

```python
plt.scatter(x, y)
plt.plot(x, Y_poly, 'r')
```

看來影響錄取率的原因不只是GRE的成績OWO
雖然想必這個預測是不準的，但還是試著測試一下

```python
from sklearn.model_selection import train_test_split
regr = LinearRegression()
```

```python
x = np.array(x)
X = x.reshape(len(x), 1)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 1023)
regr.fit(x_train, y_train)
Y_pred = regr.predict(x_test)
plt.scatter(x_test, y_test)
plt.plot(x_test, Y_pred, 'r')
```

雖然已經改為non-linear，但看來是需要多變數分析的，期待未來學習到的技巧能更準確的預測結果
