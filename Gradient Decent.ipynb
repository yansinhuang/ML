```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(-3,3,0.001)
def Loss(x):
    return x**5 + x**4 - 4*x**3 - 5*x**2 - 6*x +7 
def dLoss(x):
    return 5*x**4 + 4*x**3 - 12*x**2 - 10*x - 42
plt.plot(x,Loss(x))
```

```python
var = np.random.uniform(-3,3)
lr = 0.001
```

```python
for i in range(1000):
    var -= lr*dLoss(var)
    plt.scatter(var, Loss(var))
plt.plot(x,Loss(x))
```
